```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Loading 10% of full dataset with 17 features.

data=pd.read_csv('bank.csv',sep=';')

df=pd.DataFrame(data)

df.head()
```

```python
df.job.value_counts()
```

```python
# df1 = dropped rows where jobs were 'unknown','student','housemaid'
# shape reduced to 4287 from 4521 . deleted 234 rows
 
df1=df[(df.job != 'unknown') & (df.job != 'student') & (df.job != 'housemaid')]

df1.job.value_counts()
```

```python
df1.head()
```

```python
df1.marital.value_counts()
```

```python
df1.education.value_counts()

# df2 = dropped rows where education was 'unknown'
# shape reduced to 4136 from 4287. deleted 151 rows

df2=df1[df1.education != 'unknown']

df2.education.value_counts()
```

```python
df2['default']=df2['default'].map({'no':0,'yes':1})

df2.default.value_counts()

#dropping default ... makes no sense 

df3=df2.drop(['default'],axis=1)
```

```python
df3.head()
```

```python
df3.shape
```

```python
df3['housing']=df3['housing'].map({'no':0,'yes':1})
```

```python
df3['loan']=df3['loan'].map({'no':0,'yes':1})
```

```python
contact_dummies=pd.get_dummies(df3.contact,drop_first=True)
```

```python
contact_dummies.head(3)
```

```python
df3.duration.describe()
```

```python
df3.head()
```

```python
df3.poutcome.value_counts()
```

```python
df3['y'].value_counts()
```

```python
job_dummies=pd.get_dummies(df3.job,drop_first=True)

marital_dummies=pd.get_dummies(df3.marital,drop_first=True)

education_dummies=pd.get_dummies(df3.education,drop_first=True)

df3.head()
```

```python
df4=pd.concat([df3,job_dummies,contact_dummies,marital_dummies,education_dummies],axis=1)

df4.head()
```

```python
df4=df4.drop(['job','marital','education','contact','month'],axis=1)
df4.head()
```

```python
df4.drop('poutcome',axis=1,inplace=True)
```

```python
X=df4.drop(['y'],axis=1)
y=df4.y
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
xx = sc.fit_transform(X)
X = pd.DataFrame(xx,columns=X.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
yy = le.fit_transform(y)
y = pd.DataFrame(yy,columns=['y'])
y.head(3)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=12)

X_train.head(3)
```

```python
classifier = LogisticRegression()

from sklearn.model_selection import GridSearchCV
parameters={'penalty':['l1','l2','elasticnet'],'C':[0.1,0.5,1,2,3,4,5,6,10,20,30,40,50,100],'max_iter':[50,100,150,200,250,300]}

classifier_regressor = GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(X_train,y_train)
```

```python
classifier_regressor.best_params_
```

```python
classifier_regressor.best_score_
```

```python
y_pred=classifier_regressor.predict(X_test)

from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
probs = classifier_regressor.predict_proba(X_test)
preds = probs[:,1]
```

```python
from sklearn.metrics import roc_auc_score

scr=roc_auc_score(y_test,preds)

print(scr)
```

```python
from catboost import CatBoostClassifier
cat = CatBoostClassifier(learning_rate=0.001)
cat.fit(X_train,y_train)
```

```python
y_pred_cat = cat.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
cat_score=accuracy_score(y_pred_cat,y_test)
print(cat_score)
```

```python
from xgboost import XGBClassifier
xg = XGBClassifier()
xg.fit(X_train,y_train)
```

```python
y_pred_xg = xg.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
xg_score=accuracy_score(y_pred_xg,y_test)
print(xg_score)
```

```python
pd.options.display.float_format = '{:,.3f}'.format
score_df = pd.DataFrame({'Classifier':['Logistic Regression','CatBoost Classifier','Xgboost Classifier'],'Score':[score,cat_score,xg_score]})
score_df
```
