```python
columns=list(df_wo_corr_ft.columns)

columns.pop()

columns
```

```python
def outliers(df,ft):
    Q1 = df[ft].quantile(0.2)
    Q3 = df[ft].quantile(0.8)
    #print('q1',Q1,'q3',Q3)
    IQR = Q3-Q1
    
    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR
    
    #print('low',lower,'up',upper)
    
    ls = df.index [ (df[ft]< lower) | (df[ft] > upper) ]
    
    return ls
```

```python
index_list=[]

for feature in columns:
    index_list.extend(outliers(df_wo_corr_ft,feature))
    
print(len(index_list))
```

```python
df_wo_outliers_and_corr_ft = df_wo_corr_ft.drop(index_list,axis=0)

df_wo_outliers_and_corr_ft.shape
```

```python
df_wo_outliers_and_corr_ft
```

```python
X = df_wo_outliers_and_corr_ft.drop(['Class','Bean ID'],axis=1)
y = df_wo_outliers_and_corr_ft.Class
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV

parameters={'penalty':['l1','l2','elasticnet'],'C':[0.1,0.5,1,2,3,4,5,6,10,20,30,40,50,100,200,300],'max_iter':[50,100,150,200,250,300]}

classifier_regressor = GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(X_train,y_train)
```

```python
classifier_regressor.best_params_
```

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
y = to_categorical(y,dtype='uint8')
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential,layers

model = Sequential([
    keras.layers.Dense(100,input_shape=(11,),activation='relu'),
    keras.layers.Dense(50,activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(6,activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=100)
```

```python
model.evaluate(X_test,y_test)
```
