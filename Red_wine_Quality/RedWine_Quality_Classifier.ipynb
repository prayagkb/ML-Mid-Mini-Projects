```python
def outliers(df,ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    #print('q1',Q1,'q3',Q3)
    IQR = Q3-Q1
    
    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR
    
    #print('low',lower,'up',upper)
    
    ls = df.index [ (df[ft]< lower) | (df[ft] > upper) ]
    
    return ls
```

```python
index_list=[]

for feature in ['fixed acidity','volatile acidity','citric acid','total sulfur dioxide','pH','sulphates','alcohol']:                #### To be changed --> only remove outliers from price_per_sqft.
    index_list.extend(outliers(df,feature))
    
print(len(index_list))
```

```python
df1=df.drop(index_list,axis=0)

df2=df1.drop(['Id'],axis=1)
```

```python
X=df2.drop(['quality'],axis=1)
y=df2.quality
```
